```python
df = pd.read_csv('data.csv')
```

```python
q10=df['Q10'].str.split(",",n=20,expand=True)
q10m=q10.melt(var_name='columns', value_name='index')
result=q10m['index'].value_counts().reset_index(name='q10')
result.head(2)
```

```python
def freq(question,output):
    q=df[question].str.split(",",n=26,expand=True)
    qm=q.melt(var_name='columns', value_name='index')
    output=qm['index'].value_counts().reset_index(name=output)
    return output
    
prior_commute=freq('Q10','prior_commute');
prior_errands=freq('Q12','prior_errands');
prior_social=freq('Q13','prior_social');
prior_airport=freq('Q11','prior_airport');
current_commute=freq('Q16','current_commute');
current_errands=freq('Q17','current_errands');
current_social=freq('Q18','current_social');
current_airport=freq('Q19','current_airport');
```

```python
survey=pd.merge(prior_commute,prior_errands,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,prior_social,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,prior_airport,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,current_commute,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,current_errands,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,current_social,left_on=['index'],right_on=['index'],how='outer')
survey=pd.merge(survey,current_airport,left_on=['index'],right_on=['index'],how='outer')

survey.to_csv('transportation habit')
survey
```
